```julia
f_elu(x, c) = elu(x, 0.05) + c
function plot_m(x)
    m = x / 500 + 1
    return f_elu(m, 1.0)
end

function plot_σ(x)
    σ = x / 500 + 1
    return f_elu(σ, 2.0)
end
```
