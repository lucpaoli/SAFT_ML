```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

plt.rc('font', family='DejaVu Serif')

# Define the ELU function
def elu(x, c, alpha=0.05):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1)) + c

# Define the plot_m function based on the provided Julia code
def plot_m(x):
    m = x + 1
    return elu(m, 1.0)

# Define the plot_σ function based on the provided Julia code
def plot_σ(x):
    σ = x + 1
    return elu(σ, 2.0)

def plot_lamr(x):
    σ = x + 1
    return 10 * elu(σ, 1.0)

# Generate a range of x-values for plotting
x_values = np.linspace(-5, 5, 600)

# Calculate y-values for m and σ
m_values = [plot_m(x) for x in x_values]
σ_values = [plot_σ(x) for x in x_values]
lamr_values = [plot_lamr(x) for x in x_values]

# Create the plot
# turn grid off 
# plt.rcParams["axes.grid"] = False

plt.figure(figsize=(8, 6), dpi=400)
plt.grid(False)
# axs[1].grid(False)

plt.plot(x_values, m_values, label='$m$', color="#5dc90a", lw=3.0)
plt.plot(x_values, σ_values, label='$\sigma$', color="#08298a", lw=3.0)
plt.plot(x_values, lamr_values, label='$\lambda_r$', color="#C40233", lw=3.0)

# Add labels and a legend
plt.xlabel('X', fontsize=30)
plt.ylabel('y', fontsize=30)
plt.xlim(-2, 2)
plt.ylim(-5, 45)
plt.legend(fontsize=24)
plt.axhline(0, color='black', linewidth=2.0)
plt.axvline(0, color='black', linewidth=2.0)
# plt.tick_params(axis='both', which='major', labelsize=14)
plt.tick_params(axis='both', which='major', labelsize=24)

plt.savefig('elu.png', dpi=400, bbox_inches='tight')
plt.show()
```
